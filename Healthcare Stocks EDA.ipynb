# UNH

```python
x = 'UNH'
def close(x):
    whole[x]['Adj Close'].plot(figsize=(10, 7))
    plt.title("Adjusted Close Price of " + str(x), fontsize=16)
    # Define the labels for x-axis and y-axis
    plt.ylabel('Price', fontsize=14)
    plt.xlabel('Year', fontsize=14)
    # Plot the grid lines
    plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
    plt.show()
print(close(x))
```

```python
def percentchange(x):
    whole[str(x) +' % Change'] = whole[x]['Adj Close'].pct_change()*100
    print(whole)
print(percentchange(x))
```

```python
def plot(x):
    plt.title(str(x) + ' % Change')
    whole[str(x)+' % Change'].plot(figsize = (12, 6), fontsize = 12)
print(plot(x))
```

```python
def dailyreturns(x):
    whole[str(x)+' % Change'].hist(bins = 50, figsize = (10,5)) 
    plt.title('Frequency / Daily returns')
    plt.xlabel('Daily returns')
    plt.ylabel('Frequency')
    plt.show()
print(dailyreturns(x))
```

```python
# Statistics 
print(whole[str(x)+ ' % Change'].mean())
print(whole[str(x)+ ' % Change'].max())
print(whole[str(x)+ ' % Change'].min())
```

```python
# The average daily returns are less than 1 %. The greatest percentage was had was 12.79%, however, the worst is -17.27%. 
```

```python
# Outliers 
x = 'UNH'
def outliers(x):
    q1=whole[x]['Adj Close'].quantile(.25)
    q3=whole[x]['Adj Close'].quantile(.75)
    lfence=q1-1.5*(q3-q1)
    ufence=q3+1.5*(q3-q1)
    condition=(whole[x]['Adj Close']<lfence)|(whole[x]['Adj Close']>ufence)
    iqr_outliers=whole[x]['Adj Close'][condition]
    print(iqr_outliers)
    print(str(x), 'has', len(iqr_outliers.values),'outliers.')
    
print(outliers(x))
```

```python
x = 'JNJ'
print(outliers(x))
x = 'PFE'
print(outliers(x))
x = 'LLY'
print(outliers(x))
x = 'ABBV'
print(outliers(x))
x = 'NVO'
print(outliers(x))
x = 'MRK'
print(outliers(x))
x = 'TMO'
print(outliers(x))
x = 'AZN'
print(outliers(x))
x = 'ABT'
print(outliers(x))
```

```python
# UNH, PFE, LLY, ABBV, and NVO all have outliers. 
# The rest are pretty consistent in a 5 year time period. 
```

```python
# Which company had the greatest change of closing price (min to max)? 

x = ['UNH','JNJ','PFE','LLY','ABBV','NVO','MRK','TMO','AZN','ABT']
for i in x:
    print(i, whole[i]['Adj Close'].max() - whole[i]['Adj Close'].min())
    
# TMO (Thermo Fisher Scientific Inc.) seems to have the greatest change of closing price. 
```

What are the highest and lowest points of each company? Which companies benefited the most during the five year period? Which companies did not benefit much during the five year period? 

```python
x = ['UNH','JNJ','PFE','LLY','ABBV','NVO','MRK','TMO','AZN','ABT']
for i in x:
    print(i, ' first adjusted closing price is', whole[i]['Adj Close'].min())
    print(i, ' final adjusted closing price is', whole[i]['Adj Close'].max())
    print(whole[i]['Adj Close'].plot())


    
    
```

All companies started out below 200. Some had great changes, while other had little change. 

```python
for i in x:
    percentage = ((whole[i]['Adj Close'].max() - whole[i]['Adj Close'].min())/(whole[i]['Adj Close'].max()))*100
    print(i,percentage)
```

As mentioned before TMO has benefited the most at 79 % change, while the least benefited is JNJ with 44 %.
